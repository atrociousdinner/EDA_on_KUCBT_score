```python
male_suffixes = []
with open('./suffixes/male_suffixes.txt', 'r') as f:
    for line in f:
        male_suffixes.extend(line.strip().split())
    
female_suffixes = []
with open('./suffixes/female_suffixes.txt', 'r') as f:
    for line in f:
        female_suffixes.extend(line.strip().split())
        
male_suffixes
```

```python
name_gender_map = pd.read_csv('2022-23_name_gender.csv')
```

```python
name_gender_map
```

```python
name_gender_dict = dict(zip(name_gender_map['First_name'], name_gender_map['Gender']))
```

```python
df_2024
```

```python
df_2024['Gender'] = df_2024['First_name'].map(name_gender_dict)
```

```python
df_2024
```

```python
df_2024[~df_2024['Gender'].isna()]
```

```python
df_missing_gender = df_2024[df_2024['Gender'].isna()]
df_missing_gender
```

```python
gender_with_suffix = df_2024['Gender'].isna()
```

```python
def assign_gender_by_suffix(name):
    for suffix in male_suffixes:
        if name.endswith(suffix):
            name_gender_dict[name] = 'm'
            return 'm'
    for suffix in female_suffixes:
        if name.endswith(suffix):
            name_gender_dict[name] = 'f'
            return 'f'
    return None
    
df_2024.loc[gender_with_suffix, 'Gender'] = df_2024.loc[gender_with_suffix, 'First_name'].apply(assign_gender_by_suffix)
```

```python
df_2024[df_2024['Gender'].isna()]
```

```python
for index, row in df_2024.iterrows():
    first_name = row['First_name']
    if first_name not in name_gender_dict:
        while True:
            gender = input(f"Enter gender for {first_name} (m/f): ")
            if gender in ['f','m']:
                name_gender_dict[first_name] = gender
                df_2024.loc[index, 'Gender'] = gender
                break
```

```python
df_2024[df_2024['Gender'].isna()]

for index, row in df_2024.iterrows():
    first_name = row['First_name']
    if pd.isna(row['Gender']):
        while True:
            gender = input(f"Enter gender for {first_name} (m/f): ")
            if gender in ['f','m']:
                name_gender_dict[first_name] = gender
                df_2024.loc[index, 'Gender'] = gender
                break 
```
