```python
df = df.drop('Unnamed: 0', axis = 1).copy()
```

```python
name_gender_map = {} #first_name -> gender
first_name_list = []
for index, row in df.iterrows():
    first_name = row['Name'].split(' ')[0]
    if first_name not in name_gender_map:
        while True:
            gender = input(f"Enter gender for {first_name} (m/f): ")
            if gender in ['f','m']:
                name_gender_map[first_name] = gender
                break
```
